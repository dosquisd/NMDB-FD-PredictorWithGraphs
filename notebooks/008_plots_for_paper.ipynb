```python
import matplotlib.pyplot as plt

from utils import (
    FIGDIR,
    DistanceTransformation,
    Normalizer,
    encode_variables_to_filename,
    get_dataset_filename,
    plot_graphs_result,
    setup_plotting,
)

# setup_plotting()
```

```python
event_filename = "all.txt"
use_threshold = imput_data = False
df = get_dataset_filename(event_filename, imput_data, use_threshold)

normalizer = Normalizer.Z_SCORE
transformation = DistanceTransformation.EXPONENTIAL
event_date = "2012-03-08"

plot_df = df[
    (df["normalization"] == normalizer.value)
    & (df["transformation"] == transformation.value)
    & (df["event_date"] == event_date)
]

filename_encoded = encode_variables_to_filename(
    event_filename, imput_data, use_threshold
).removesuffix(".csv")
filename_encoded += (
    f"_transformation-{transformation.value}_normalizer-{normalizer.value}"
)

format = "png"

fig = plot_graphs_result(
    plot_df,
    ncols_multiplier=9.0,
    altitude_normalizer=Normalizer.Z_SCORE,
    cmap=plt.cm.Spectral_r,
)
fig.savefig(FIGDIR / f"{filename_encoded}-{event_date}-fixed.{format}", format=format)

fig = plot_graphs_result(
    plot_df,
    ncols_multiplier=9.0,
    fixed_node_size=500,
    cmap=plt.cm.Spectral_r,
)
fig.savefig(FIGDIR / f"{filename_encoded}-{event_date}-altitude.{format}", format=format)
```
